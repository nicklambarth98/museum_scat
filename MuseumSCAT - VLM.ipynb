## Calling VLM and Parsing training set
---
In this section I will call a pre-trained VLM to classify the text. This method was chosen for this notebook as the training set is rather small and it is likely not an appropriate quantity of data to train such a large model on.  

Additionally, it is a good baseline to combine heuristic methods against as well as providing an opportuntity to test a classification confidence threshold (inspired by the work in [this paper by Geifman and El-Yaniv](https://proceedings.neurips.cc/paper_files/paper/2017/file/4a8423d5e91fda00bb7e46540e2b0cf1-Paper.pdf))

There will be two VLM runs in this notebook, a single estimation per sample and 3 estimations per image with a confidence threshold for submission. Both methods will be compared and the submission with the best AURC will be submitted. The threshold is to ensure that the model does not submit a specific classification for something that it is not sure about.  

In this instance the cost of being confident about something wrong may lead to an incorrect label interpretation which would cause more disruption than the model just saying "i don't understand please help" on some samples.

### Creating Helper and Pre-processing functions

````python
def check_ollama_running() -> bool:
    """
    Ping the Ollama server. Returns True if reachable.
    """
    try:
        ollama.list()   # lightweight endpoint — lists downloaded models
        print('✓ Ollama server is running')
        return True
    except Exception as e:
        print(f'✗ Ollama not reachable: {e}')
        print('  → Start it with: ollama serve')
        return False
````

````python
def ensure_model_pulled(model_name: str) -> bool:
    """
    Check if the model is already downloaded; pull it if not.
    """
    available = [m.model for m in ollama.list().models]
    # Normalise ollama list 
    short_names = [m.split(':')[0] for m in available]

    if model_name.split(':')[0] in short_names:
        print(f'✓ Model {model_name!r} already available')
        return True

    print(f'  Pulling {model_name!r}')
    ollama.pull(model_name)
    print(f'✓ Model {model_name!r} pulled successfully')
    return True


# Run at startup
if check_ollama_running():
    ensure_model_pulled(MODEL_NAME)
````

````python
def load_and_preprocess_image(
    img_path: Path,
    max_dim: int = 1024,
    enhance_contrast: bool = True,
) -> Image.Image:
    """
    Load an image and apply lightweight preprocessing:
    - Resize so the longest side <= max_dim (preserving aspect ratio)
    - CLAHE contrast enhancement for faded labels

    Note: most Ollama vision models internally resize to ~336–448px anyway,
    so very high resolutions are wasted compute. 1024px is a good ceiling.
    """
    img = Image.open(img_path).convert('RGB')

    # Resize
    w, h = img.size
    if max(w, h) > max_dim:
        scale = max_dim / max(w, h)
        img = img.resize((int(w * scale), int(h * scale)), Image.LANCZOS)

    # CLAHE contrast enhancement — helps with faded/aged labels
    if enhance_contrast:
        arr  = np.array(img)
        lab  = cv2.cvtColor(arr, cv2.COLOR_RGB2LAB)
        l, a, b = cv2.split(lab)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        l     = clahe.apply(l)
        lab   = cv2.merge([l, a, b])
        arr   = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)
        img   = Image.fromarray(arr)

    return img
````

````python
def image_to_bytes(img: Image.Image, fmt: str = 'JPEG') -> bytes:
    """Convert a PIL Image to raw bytes for passing to ollama.chat()."""
    import io
    buf = io.BytesIO()
    img.save(buf, format=fmt)
    return buf.getvalue()


def get_image_path(image_id: str) -> Path:
    """Resolve the full path for a given image_id."""
    return IMAGE_DIR / f"{image_id}{IMAGE_EXT}"
````

### Prompt Construction

````python
SYSTEM_PROMPT = """You are an expert transcription assistant specialising in historical
Danish entomological museum specimen labels. Your task is to read label images and
extract exactly two fields.

RULES:
- Transcribe VERBATIM — copy exactly what is written, including abbreviations,
  Roman numerals, unusual punctuation, and old spellings.
- Use the Danish alphabet: æ, ø, å (NOT ö, ü, ä, ÿ — convert these to Danish equivalents).
- Roman numerals for months must be preserved (e.g. IV, VI, IX).
- If a specimen has multiple label cards, separate values with ' | '.
- If a field is absent from all labels, output NULL.
- Do NOT interpret, standardise, or add information not visible on the label.
- 'Dania' is NOT a locality — do not include it in verbatimLocality.
- Phrases like 'i kogødning' (in cow dung) are NOT localities — omit them.

OUTPUT FORMAT — respond with ONLY valid JSON, no other text:
{"verbatimDate": "...", "verbatimLocality": "...", "confidence_date": 0-100, "confidence_locality": 0-100}"""

USER_PROMPT = (
    "Transcribe the verbatimDate and verbatimLocality from this specimen label image. "
    "Output only the JSON object described in your instructions."
)


def build_messages(img_bytes: bytes) -> list[dict]:
    """
    Build the messages list for ollama.chat().
    The system prompt is passed as a separate system message;
    the image bytes go in the user message's 'images' field.
    """
    return [
        {'role': 'system', 'content': SYSTEM_PROMPT},
        {'role': 'user',   'content': USER_PROMPT, 'images': [img_bytes]},
    ]
````

### Calling Models

````python
@dataclass
class RawPrediction:
    """Holds the raw string output from a single model call."""
    image_id:    str
    raw_text:    str
    model:       str
    temperature: float
    elapsed_s:   float
    error:       Optional[str] = None
````

````python
def call_model_once(
    image_id: str,
    img_bytes: bytes,
    model: str = MODEL_NAME,
    temperature: float = 0.0,
    max_retries: int = 3,
    retry_delay: float = 5.0,
) -> RawPrediction:
    """
    Make a single model call. Retries on transient errors (e.g. model loading).
    temperature=0.0 gives deterministic output (greedy decoding).
    """
    messages = build_messages(img_bytes)

    for attempt in range(max_retries):
        try:
            t0 = time.time()
            response = ollama.chat(
                model=model,
                messages=messages,
                options={'temperature': temperature},
            )
            elapsed = time.time() - t0
            return RawPrediction(
                image_id=image_id,
                raw_text=response.message.content,
                model=model,
                temperature=temperature,
                elapsed_s=elapsed,
            )
        except Exception as e:
            if attempt < max_retries - 1:
                print(f'  Retry {attempt+1}/{max_retries} for {image_id}: {e}')
                time.sleep(retry_delay * (attempt + 1))  # exponential back-off
            else:
                return RawPrediction(
                    image_id=image_id, raw_text='',
                    model=model, temperature=temperature,
                    elapsed_s=0.0, error=str(e)
                )
````

````python
def call_model_multi_sample(
    image_id: str,
    img_bytes: bytes,
    model: str = MODEL_NAME,
    n_samples: int = N_SAMPLES,
    temperature: float = TEMPERATURE,
) -> list[RawPrediction]:
    """
    Run the model n_samples times at a given temperature.
    Used for the MC-Dropout-style confidence estimation in Submission 2.
    Returns a list of RawPrediction objects — one per sample.
    """
    return [
        call_model_once(image_id, img_bytes, model=model, temperature=temperature)
        for _ in range(n_samples)
    ]
````

### Caching responses

````python
def _cache_key(image_id: str, model: str, n_samples: int, temperature: float) -> str:
    """Deterministic cache key from call parameters."""
    raw = f"{image_id}|{model}|{n_samples}|{temperature}"
    return hashlib.md5(raw.encode()).hexdigest()


def load_from_cache(image_id: str, model: str, n_samples: int, temperature: float):
    key  = _cache_key(image_id, model, n_samples, temperature)
    path = CACHE_DIR / f"{key}.pkl"
    if path.exists():
        with open(path, 'rb') as f:
            return pickle.load(f)
    return None


def save_to_cache(image_id: str, model: str, n_samples: int,
                  temperature: float, data) -> None:
    key  = _cache_key(image_id, model, n_samples, temperature)
    path = CACHE_DIR / f"{key}.pkl"
    with open(path, 'wb') as f:
        pickle.dump(data, f)


def cached_call(
    image_id: str,
    img_bytes: bytes,
    model: str = MODEL_NAME,
    n_samples: int = 1,
    temperature: float = 0.0,
) -> list[RawPrediction]:
    """
    Wrapper around call_model_* with disk caching.
    Re-running the notebook won't re-call the model for already-processed images.
    """
    cached = load_from_cache(image_id, model, n_samples, temperature)
    if cached is not None:
        return cached

    if n_samples == 1:
        result = [call_model_once(image_id, img_bytes, model, temperature)]
    else:
        result = call_model_multi_sample(image_id, img_bytes, model, n_samples, temperature)

    save_to_cache(image_id, model, n_samples, temperature, result)
    return result
````

### Output parsing and post processing

````python
@dataclass
class ParsedPrediction:
    image_id:           str
    verbatimDate:       str
    verbatimLocality:   str
    confidence_date:    float        # 0–1
    confidence_locality:float        # 0–1
    parse_error:        Optional[str] = None


# Umlaut mapping — convert legacy German umlauts to Danish equivalents
_UMLAUT_MAP = str.maketrans({'ö': 'ø', 'Ö': 'Ø', 'ü': 'y', 'Ü': 'Y',
                              'ä': 'æ', 'Ä': 'Æ', 'ÿ': 'y'})

# Phrases to strip from locality (not places)
_LOCALITY_STRIP = [
    r'\bDania\b',                          # collection name, not a place
    r'i\s+ko[gj][øo]dning',               # 'in cow dung'
    r'i\s+ko[gj][øo]rning',
    r'Mus\.?\s*L[øo]v\.?',               # museum reference
]
````

````python
def normalise_field(text: str) -> str:
    """Apply umlaut conversion and strip known non-locality phrases."""
    if not text or text.upper() == 'NULL':
        return 'NULL'
    text = text.translate(_UMLAUT_MAP)
    return text.strip()


def clean_locality(text: str) -> str:
    """Remove known non-place phrases from a locality string."""
    if not text or text.upper() == 'NULL':
        return 'NULL'
    for pattern in _LOCALITY_STRIP:
        text = re.sub(pattern, '', text, flags=re.IGNORECASE).strip(', |').strip()
    return text.strip() or 'NULL'
````

````python
def parse_raw_response(image_id: str, raw_text: str) -> ParsedPrediction:
    """
    Extract JSON from model output and build a ParsedPrediction.
    Handles common model failure modes:
    - JSON wrapped in markdown code fences
    - Trailing commentary after the JSON object
    - Missing fields
    """
    # Strip markdown code fences if present
    text = re.sub(r'```(?:json)?\s*', '', raw_text).strip()

    # Extract first {...} block
    match = re.search(r'\{[^{}]+\}', text, re.DOTALL)
    if not match:
        return ParsedPrediction(
            image_id=image_id,
            verbatimDate='NULL', verbatimLocality='NULL',
            confidence_date=0.0, confidence_locality=0.0,
            parse_error=f'No JSON found in: {raw_text[:100]}'
        )

    try:
        data = json.loads(match.group())
    except json.JSONDecodeError as e:
        return ParsedPrediction(
            image_id=image_id,
            verbatimDate='NULL', verbatimLocality='NULL',
            confidence_date=0.0, confidence_locality=0.0,
            parse_error=f'JSON decode error: {e}'
        )

    date     = normalise_field(str(data.get('verbatimDate',     'NULL')))
    locality = clean_locality(normalise_field(str(data.get('verbatimLocality', 'NULL'))))
    conf_d   = float(data.get('confidence_date',     50)) / 100.0
    conf_l   = float(data.get('confidence_locality', 50)) / 100.0

    return ParsedPrediction(
        image_id=image_id,
        verbatimDate=date,
        verbatimLocality=locality,
        confidence_date=conf_d,
        confidence_locality=conf_l,
    )
````

````python
def aggregate_multi_sample(
    image_id: str,
    raw_preds: list[RawPrediction],
) -> ParsedPrediction:
    """
    Aggregate N parsed predictions into one by majority vote.
    Cross-sample agreement is used as the confidence signal
    (mirrors MC-Dropout in the Geifman & El-Yaniv paper).

    For each field:
    - Take the most common output across N runs
    - Confidence = fraction of runs that agreed on that output
    """
    parsed = [parse_raw_response(image_id, r.raw_text) for r in raw_preds
              if not r.error]

    if not parsed:
        return ParsedPrediction(
            image_id=image_id,
            verbatimDate='NULL', verbatimLocality='NULL',
            confidence_date=0.0, confidence_locality=0.0,
            parse_error='All samples errored'
        )

    def majority_vote(values: list[str]) -> tuple[str, float]:
        from collections import Counter
        counts = Counter(values)
        winner, count = counts.most_common(1)[0]
        return winner, count / len(values)

    dates      = [p.verbatimDate     for p in parsed]
    localities = [p.verbatimLocality for p in parsed]

    best_date, conf_d = majority_vote(dates)
    best_loc,  conf_l = majority_vote(localities)

    return ParsedPrediction(
        image_id=image_id,
        verbatimDate=best_date,
        verbatimLocality=best_loc,
        confidence_date=conf_d,
        confidence_locality=conf_l,
    )
````

### Scoring Utility

````python
# Punctuation normalisation as specified by the competition metric
_PUNCT_NORM = re.compile(r'[.,\-·\s]+')


def normalise_for_scoring(text: str) -> str:
    """Apply competition punctuation normalisation before comparison."""
    if text.upper() == 'NULL':
        return 'NULL'
    return _PUNCT_NORM.sub(' ', text).strip().lower()


def score_prediction(pred: str, gold: str) -> float:
    """
    Returns 1.0 for exact match (after normalisation), else 0.0.
    Extend to character-level partial credit if the competition allows it.
    """
    return float(normalise_for_scoring(pred) == normalise_for_scoring(gold))
````

````python
def score_dataframe(df_pred: pd.DataFrame, df_gold: pd.DataFrame,
                    id_col: str = IMAGE_ID_COL) -> dict:
    """
    Score a full prediction DataFrame against ground truth.
    Returns a dict with per-field and overall accuracy.
    """
    merged = df_pred.merge(df_gold[[id_col, 'verbatimDate', 'verbatimLocality']],
                           on=id_col, suffixes=('_pred', '_gold'))

    merged['score_date'] = merged.apply(
        lambda r: score_prediction(r['verbatimDate_pred'], r['verbatimDate_gold']), axis=1)
    merged['score_loc']  = merged.apply(
        lambda r: score_prediction(r['verbatimLocality_pred'], r['verbatimLocality_gold']), axis=1)
    merged['score_both'] = (merged['score_date'] + merged['score_loc']) / 2

    return {
        'date_accuracy':     merged['score_date'].mean(),
        'locality_accuracy': merged['score_loc'].mean(),
        'overall':           merged['score_both'].mean(),
        'n':                 len(merged),
        'df':                merged,
    }
````

### Full pipeline Orchestration

````python
def run_pipeline(
    df: pd.DataFrame,
    model: str = MODEL_NAME,
    n_samples: int = 1,
    temperature: float = 0.0,
    enhance_contrast: bool = True,
) -> pd.DataFrame:
    """
    Run the full pipeline over a DataFrame of specimens.

    Parameters
    ----------
    df          : DataFrame with an IMAGE_ID_COL column
    n_samples   : 1 = deterministic single call (Submission 1)
                  3+ = multi-sample for confidence estimation (Submission 2)
    temperature : 0.0 for deterministic; 0.7 for multi-sample diversity

    Returns
    -------
    DataFrame with columns: image_id, verbatimDate, verbatimLocality,
                             confidence_date, confidence_locality
    """
    results = []

    for _, row in tqdm(df.iterrows(), total=len(df), desc='Processing specimens'):
        image_id = str(row[IMAGE_ID_COL])
        img_path = get_image_path(image_id)

        # ── Load & preprocess
        try:
            img  = load_and_preprocess_image(img_path, enhance_contrast=enhance_contrast)
            img_bytes = image_to_bytes(img)
        except FileNotFoundError:
            results.append(ParsedPrediction(
                image_id=image_id,
                verbatimDate='NULL', verbatimLocality='NULL',
                confidence_date=0.0, confidence_locality=0.0,
                parse_error='Image not found'
            ))
            continue

        # ── Call model (with caching)
        raw_preds = cached_call(image_id, img_bytes, model, n_samples, temperature)

        # ── Parse & aggregate
        if n_samples == 1:
            parsed = parse_raw_response(image_id, raw_preds[0].raw_text)
        else:
            parsed = aggregate_multi_sample(image_id, raw_preds)

        results.append(parsed)

    # Convert to DataFrame
    return pd.DataFrame([
        {
            IMAGE_ID_COL:          r.image_id,
            'verbatimDate':        r.verbatimDate,
            'verbatimLocality':    r.verbatimLocality,
            'confidence_date':     r.confidence_date,
            'confidence_locality': r.confidence_locality,
            'parse_error':         r.parse_error,
        }
        for r in results
    ])
````

## Submission 1 - Single VLM Estimation

````python
# ── Validate on training set first
train_df = pd.read_csv(TRAIN_CSV)
val_preds = run_pipeline(train_df, n_samples=1, temperature=0.0)

scores = score_dataframe(val_preds, train_df)
print(f"Validation — date: {scores['date_accuracy']:.3f}  "
      f"locality: {scores['locality_accuracy']:.3f}  "
      f"overall: {scores['overall']:.3f}  (n={scores['n']})")

# Inspect failures
failures = scores['df'][scores['df']['score_both'] < 1.0]
print(f"\nFailed predictions: {len(failures)}")
failures[[
    IMAGE_ID_COL,
    'verbatimDate_pred', 'verbatimDate_gold',
    'verbatimLocality_pred', 'verbatimLocality_gold'
]].head(10)
````

## Submission 2 - Multiple VLM Estimations and Selective Classifications

````python
# ── Run multi-sample on training set to calibrate threshold
val_preds_ms = run_pipeline(train_df, n_samples=N_SAMPLES, temperature=TEMPERATURE)
val_scored   = score_dataframe(val_preds_ms, train_df)
val_merged   = val_scored['df']

# Composite confidence = mean of date and locality confidence
val_merged['confidence'] = (
    val_preds_ms.set_index(IMAGE_ID_COL)['confidence_date'] +
    val_preds_ms.set_index(IMAGE_ID_COL)['confidence_locality']
).reindex(val_merged[IMAGE_ID_COL]).values / 2


# ── Plot coverage–risk curve
import matplotlib.pyplot as plt

thresholds = np.linspace(0, 1, 50)
coverages, risks = [], []

for theta in thresholds:
    accepted  = val_merged[val_merged['confidence'] >= theta]
    coverage  = len(accepted) / len(val_merged)
    risk      = 1 - accepted['score_both'].mean() if len(accepted) > 0 else np.nan
    coverages.append(coverage)
    risks.append(risk)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(coverages, risks, marker='o', markersize=3)
ax.set_xlabel('Coverage (fraction of specimens predicted)')
ax.set_ylabel('Risk (error rate on accepted predictions)')
ax.set_title('Coverage–Risk Curve (Geifman & El-Yaniv framework)')
ax.invert_xaxis()  # low coverage (high threshold) on the right
plt.tight_layout()
plt.show()
````

````python
# ── Pick threshold — look for the elbow in the coverage–risk curve
# A reasonable heuristic: find the lowest threshold where risk < target
TARGET_RISK = 0.15   # adjust based on your curve

best_theta = None
for theta, cov, risk in zip(thresholds, coverages, risks):
    if not np.isnan(risk) and risk < TARGET_RISK:
        best_theta = theta
        break  # first (lowest threshold) that meets the risk target

print(f"Selected threshold θ* = {best_theta:.2f}")
accepted_val = val_merged[val_merged['confidence'] >= best_theta]
print(f"Coverage: {len(accepted_val)/len(val_merged):.1%}  "
      f"Risk on accepted: {1 - accepted_val['score_both'].mean():.3f}")
````

````python
# ── Run multi-sample on test set
test_preds_ms = run_pipeline(test_df, n_samples=N_SAMPLES, temperature=TEMPERATURE)

test_preds_ms['confidence'] = (
    test_preds_ms['confidence_date'] + test_preds_ms['confidence_locality']
) / 2

# ── Apply selective classification: fallback to Submission 1 output for low-confidence
# Merge multi-sample predictions with Submission 1 predictions
merged_test = test_preds_ms.merge(
    test_preds_s1[[IMAGE_ID_COL, 'verbatimDate', 'verbatimLocality']],
    on=IMAGE_ID_COL, suffixes=('_ms', '_s1')
)

def select_prediction(row, theta: float) -> tuple[str, str]:
    """
    Accept multi-sample prediction if confidence >= theta.
    Otherwise fall back to Submission 1 (single deterministic call).
    Both-NULL is the final fallback if neither is confident.
    """
    if row['confidence'] >= theta:
        return row['verbatimDate_ms'], row['verbatimLocality_ms']
    else:
        # Fallback: use the single deterministic run from Submission 1
        return row['verbatimDate_s1'], row['verbatimLocality_s1']

merged_test[['verbatimDate', 'verbatimLocality']] = merged_test.apply(
    lambda r: pd.Series(select_prediction(r, best_theta)), axis=1
)

sub2 = merged_test[[IMAGE_ID_COL, 'verbatimDate', 'verbatimLocality']]
sub2.to_csv(OUTPUT_DIR / 'submission_2_selective.csv', index=False)
print(f"Submission 2 written: {len(sub2)} rows")

accepted_count = (merged_test['confidence'] >= best_theta).sum()
print(f"Accepted (multi-sample): {accepted_count}  "
      f"Fell back to S1: {len(merged_test) - accepted_count}")
````
